```python
# Reshaping, stacking, squeezing, unsqueezing tensors
import torch
x= torch.arange(1,11)
x, x.shape
```

```python
# add an extra dimension
# x_reshaped= x.reshape(1,10)
# x_reshaped= x.reshape(10,1)
x_reshaped= x.reshape(5,2)
x_reshaped, x_reshaped.shape
```

```python
# change the view, shares the same memory as the original tensor
z= x.view(2,5)
z, z.shape
```

```python
#changing the z changes the x
z[0][1]=5
z,x
```

```python
# stack tensors on top of each others
# dim =1 horizontally, dim=2 vertically
x_stacked= torch.stack([x, x], dim=1)
x_stacked
```
