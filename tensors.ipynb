```python
# Reshaping, stacking, squeezing, unsqueezing tensors
import torch
x= torch.arange(1,11)
x, x.shape
```

```python
# add an extra dimension
# x_reshaped= x.reshape(1,10)
# x_reshaped= x.reshape(10,1)
x_reshaped= x.reshape(5,2)
x_reshaped, x_reshaped.shape
```

```python
# change the view, shares the same memory as the original tensor
z= x.view(2,5)
z, z.shape
```

```python
#changing the z changes the x
z[0][1]=5
z,x
```

```python
# stack tensors on top of each others
# dim =1 horizontally, dim=2 vertically
x_stacked= torch.stack([x, x], dim=1)
x_stacked
```

```python
#squeeze- removes all single dimensions from a target tensor
x =torch.zeros(2,1,3,1)
y= torch.squeeze(x)
y.shape, x.shape, y
```

```python
#unsqueeze - adds a single dimension to a target tensor at a speciic dim(dimension)
y= torch.rand(3)
print(f"previous target: {y}")
print(f"previous target shape: {y.shape}")
z= torch.unsqueeze(y, dim=0)
print(f"\nnew target: {z}")
print(f"new target shape: {z.shape}")
f= torch.unsqueeze(y, dim=1)
print(f"\nnew target on dim 1: {f}")
print(f"new target shape: {f.shape}")
```

```python
# torch.permute- rearrange  the dimension of a target tensor in a specified order
y= torch.rand(3,2,5)
print(f"target tensor: {y}")
print(f"\ntarget tensor shape: {y.shape}")
z= torch.permute(y, (2,0,1))
print(f"\n\nnew target: {z}")
print(f"\nnew target shape: {z.shape}")
```
