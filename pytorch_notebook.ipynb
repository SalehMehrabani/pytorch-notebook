```python
import time

#time test tensor creation
start_time= time.time()
tensor1= torch.tensor([[1,-1],[2,-2]])
print(f'time to create a tensor without numpy: {time.time()-start_time}')
print(tensor1)

start_time= time.time()
tensor2= torch.tensor(np.array([[1,-1],[2,-2]]))
print(f'time to create a tensor with numpy: {time.time()-start_time}')
print(tensor2)
```

```python
# Create a range of tensors and tensors-like
one_to_ten=torch.arange(start=1,end=11, step=1)
one_to_ten
```

```python
#Creating tensors-like
ten_zeros= torch.zeros_like(one_to_ten, dtype=None)
ten_zeros
```

```python
# tensors data type
ten_zeros.dtype, one_to_ten.dtype
```

```python
# data type conversion
tens_zeros_float16= ten_zeros.type(torch.float16)
tens_zeros_float16.dtype
```

```python
# Getting information from tensors
tens_zeros_float16.dtype, tens_zeros_float16.shape, tens_zeros_float16.device
```
